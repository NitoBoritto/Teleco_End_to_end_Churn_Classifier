# Encoding

```python
# Binary encoding
binary_cols = ['gender', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']

df[binary_cols] = df[binary_cols].replace({
    'Yes': 1, 'No': 0,
    'Male': 1, 'Female': 0
}).astype(int)
```

```python
df[binary_cols].head()
```

```python
# One hot encoding
multi_cat_cols = [
    'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
    'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaymentMethod'
]

df = pd.get_dummies(df, columns = multi_cat_cols, drop_first = True)
```

```python
df.head()
```

# Cleaning

```python
# Changing dtypes
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors = 'coerce')

bool_cols = df.select_dtypes(include = 'bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
df.info()
```

```python
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna()
df.head()
```

```python
# Correlation matrix
corr_matrix = df.corr(numeric_only = True)

churn_corr = corr_matrix[['Churn']].sort_values(by = 'Churn', ascending = False)

plt.figure(figsize = (4, 8))
sns.heatmap(churn_corr,
            annot = True,
            cmap = 'coolwarm',
            vmin = -1,
            vmax = 1)
plt.title('Correlation of features with churn')
plt.tight_layout
plt.show()
```

# Multicolinearity (VIF)

```python
# Dropping "No internet service" redundancy
df['No_internet_service'] = (
    df['OnlineSecurity_No internet service'] |
    df['OnlineBackup_No internet service'] |
    df['DeviceProtection_No internet service'] |
    df['TechSupport_No internet service'] |
    df['StreamingTV_No internet service'] |
    df['StreamingMovies_No internet service']
).astype(int)

drop_cols = [col for col in df.columns if 'No internet service' in col]
df = df.drop(columns = drop_cols)

# Dropping "No phone service" redundancy
if 'MultipleLines_No phone service' in df.columns:
    df['No_phone_service'] = df['MultipleLines_No phone service'].astype(int)
    df = df.drop(columns = 'MultipleLines_No phone service')
```

```python
X = df.drop(columns = ['Churn', 'InternetService_No', 'PhoneService', 'MonthlyCharges'])
X = X.replace([np.inf, -np.inf], np.nan)
X = X.dropna()
print(X.shape)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif_data = vif_data.sort_values(by = 'VIF', ascending = False)
print(vif_data)
```

# Modeling

```python
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, cross_validate
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from skopt.searchcv import BayesSearchCV
from skopt.space import Real, Integer, Categorical
from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score, f1_score
from sklearn.preprocessing import RobustScaler
import time
```

```python
y = df['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, stratify = y, random_state = 30)

threshold = .3 # Lowering threshold (.5 by default) to boost recall

scaler = RobustScaler()

scaling_features = ['TotalCharges', 'tenure']

X_train[scaling_features] = scaler.fit_transform(X_train[scaling_features])
X_test[scaling_features] = scaler.transform(X_test[scaling_features])
```

## Logistic Regression

```python
lgr_params = {'C': Real(.1, 1, prior = 'log-uniform')}

lgr_search = BayesSearchCV(estimator = LogisticRegression(random_state = 30, max_iter = 1000, class_weight = 'balanced'),
                           search_spaces = lgr_params,
                           scoring = 'recall',
                           n_iter = 30,
                           n_jobs = -1,
                           random_state = 30,
                           cv = 5)

lgr_search.fit(X_train, y_train)
lgr_best_params = lgr_search.best_params_
print(f'Best parameters: {lgr_best_params}')
```

```python
lgr = LogisticRegression(**lgr_best_params, random_state = 30, max_iter = 1000, class_weight = 'balanced')

# Training timer
start_train = time.time()
lgr.fit(X_train, y_train)
train_time = time.time() - start_train
print(f'Training time: {train_time:.2f} seconds')


# Predicting timer
start_predict = time.time()
y_pred = lgr.predict(X_test)
y_proba = lgr.predict_proba(X_test)
predict_time = time.time() - start_predict
print(f'Predecting time: {predict_time:.2f} seconds')

# Threshhold Tuning
target_proba = y_proba[:, 1]
y_pred_30 = (target_proba >= threshold).astype(int)

# Classification Report
print(f'Classification Report (.5 Threshold):\n{classification_report(y_test, y_pred)}')
print('----------------------------------------------------------------\n')
print(f'Classification Report (.3 Threshold):\n{classification_report(y_test, y_pred_30)}')
print('Not bad but we can do better\n ')
```

# Hyperparameter tuning using Optuna

```python
import optuna
scale_pos_weights = (y_train == 0).sum() / (y_train == 1).sum()
def objective(trial):

    params = {
        'objective': 'binary:logistic',
        'random_state': 30,
        'n_jobs': -1,
        # Search spaces
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'max_depth': trial.suggest_int('max_depth', 3, 15),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'alpha': trial.suggest_float('alpha', 1e-3, 10.0, log=True),
        'lambda': trial.suggest_float('lambda', 1e-3, 10.0, log=True),
        'gamma': trial.suggest_float('gamma', 0, 5),
        # min_child_weight: higher values prevent the model from learning highly specific patterns
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'scale_pos_weight': trial.suggest_float('scale_pos_weight', scale_pos_weights * 0.8, scale_pos_weights * 2)
    }

    xgb_optuna = XGBClassifier(**params)
    xgb_optuna.fit(X_train, y_train)
    # using cross_val_score to maximize recall
    score = cross_val_score(xgb_optuna, X_train, y_train, cv = 5, scoring = 'recall', n_jobs=-1).mean()
    
    return score

# Create and Run the Study
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials = 100)

print(f"Best Recall Score: {study.best_value}")
print(f"Best Params: {study.best_params}")
```

```python
scale_pos_weights = (y_train == 0).sum() / (y_train == 1).sum()
xgb_best_params = study.best_params

xgb = XGBClassifier(**xgb_best_params,
                    random_state = 30,
                    n_jobs = -1,
                    scale_pos_weights = scale_pos_weights)

start_train = time.time()
xgb.fit(X_train, y_train)
train_time = time.time() - start_train
print(f'Training time: {train_time:.2f} seconds')

start_predict = time.time()
y_pred = xgb.predict(X_test)
y_proba = xgb.predict_proba(X_test)
predict_time = time.time() - start_predict
print(f'Predecting time: {predict_time:.2f} seconds')

target_proba = y_proba[:, 1]
y_pred_30 = (target_proba >= threshold).astype(int)

# Classification Report
print(f'Classification Report (.5 Threshold):\n{classification_report(y_test, y_pred)}')
print('----------------------------------------------------------------\n')
print(f'Classification Report (.3 Threshold):\n{classification_report(y_test, y_pred_30)}')
```

Using Optuna Helped with hyperparameter tuning and resulted in a much higher score but sacrificing percision for churners

I choose Logistic regression for its simplicity and faster training time while not loosing much recall and have a better f1 score (all rounder)
